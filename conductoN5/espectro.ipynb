```python
import pandas as pd

# Saltar las líneas de encabezado que empiezan con '#' o '@'
df = pd.read_csv(
    "espectro.DAT",
    delim_whitespace=True,
    comment='@',
    skiprows=3,
    skipfooter=3,  # Salta las últimas 3 filas
    header=None,
    names=["E_meV", "ESP", "ERR", "CANI", "CANS", "DCAN"],
    engine='python'
)

import matplotlib.pyplot as plt
%matplotlib widget
plt.figure()
plt.errorbar(
    df["E_meV"],
    df["ESP"],
    yerr=df["ERR"],
)
plt.xlabel("Energía media (meV)")
plt.ylabel("Valor del espectro")
plt.yscale("log")  # Escala logarítmica en Y
plt.xscale("log")  # Escala logarítmica en X
plt.title("Espectro con barras de error")
plt.grid(True)
plt.show()
```
