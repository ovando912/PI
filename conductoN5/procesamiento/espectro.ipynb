```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_puntos_intermedios_log_extendido(energias):
    energias = np.array(energias)
    if np.any(energias <= 0):
        raise ValueError("Todas las energías deben ser mayores a cero para el logaritmo.")

    log_energias = np.log(energias)
    
    # Puntos medios logarítmicos
    log_medios = 0.5 * (log_energias[:-1] + log_energias[1:])
    
    # Calcular el delta log típico (no constante, pero usamos el promedio)
    delta_log = np.mean(np.diff(log_energias))

    # Extender un punto antes y después en el espacio logarítmico
    log_extra_izq = log_energias[0] - delta_log/2
    log_extra_der = log_energias[-1] + delta_log/2

    # Armar el vector completo en log
    log_total = np.concatenate([[log_extra_izq], log_medios, [log_extra_der]])
    total = np.exp(log_total)

    return total

# Usá tu lista de energías
energias = lista_energias
medios_ext = calcular_puntos_intermedios_log_extendido(energias)
print(medios_ext)

# Plot para verificar
plt.figure(figsize=(8, 5))
plt.plot(energias, np.ones_like(energias), 'o', label='Originales')
plt.plot(medios_ext, np.ones_like(medios_ext), 'x', label='Intermedios + extendido', color='red')
plt.xscale('log')
plt.ylim(0.99, 1.02)
plt.legend()
plt.title("Puntos intermedios logarítmicos con extensión a ambos lados")
plt.grid(True, which='both')
plt.show()
```
