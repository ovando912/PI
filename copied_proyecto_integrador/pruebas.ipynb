```python
# np.random.seed(7)
a = np.random.rand(25)
b = np.random.rand(25) + a

def calificador(a: np.ndarray, b: np.ndarray):
    min = np.sort(a)
    # print(min)
    max = np.sort(b)
    # print(max)
    c = (a+b)/2
    med = np.sort(c)
    # print(med)

    d = b-a

    largos = np.sort(d)

    calification = np.zeros(len(a))
    # print("med ", c)
    # print(med)
    for i, (x, y, z,w) in enumerate(zip(a, b, c,d)):
        # Find the index of x in min_sorted, y in max_sorted, and z in med_sorted
        x_index = np.where(min == x)[0][0]  # Get the first occurrence
        y_index = np.where(max == y)[0][0]
        z_index = np.where(med == z)[0][0]
        w_index = np.where(largos == w)[0][0]
        # print(z_index)

        # Add these indices to calification
        # calification[i] =- (x_index + 1.3*y_index + z_index + 0.5*w_index)
        calification[i] =- (x_index)

    return calification

califica = calificador(a, b)




# Create a figure with two subplots horizontally
axes = plt.subplots(3, 1, figsize=(12, 6))[1]

for i in range(len(a)):
    if i % 2 == 0:
        axes[0].plot([i, i], [a[i], b[i]], color='blue', lw=2, marker='o')  # Plot the box from min to max
    else:
        axes[0].plot([i, i], [a[i], b[i]], color='red', lw=2, marker='o')
axes[0].set_xticks(np.arange(len(a)))
axes[0].set_xticklabels(range(len(a)),rotation=90)
axes[0].grid()
axes[0].set_yscale('log')
axes[0].set_title('Min to Max')
axes[0].set_xlabel('Index')



c = np.random.rand(25)
d = np.random.rand(25) + c

# califica += calificador(c, d)
aux = np.sort(califica)
for i,x in enumerate(califica):
    califica[i] = np.where(aux == x)[0][0]

for i in range(len(a)):
    if i % 2 == 0:
        axes[1].plot([i, i], [c[i], d[i]], color='blue', lw=2, marker='o')  # Plot the box from min to max
    else:
        axes[1].plot([i, i], [c[i], d[i]], color='red', lw=2, marker='o')
axes[1].set_xticks(np.arange(len(a)))
axes[1].set_xticklabels(range(len(a)),rotation=90)
axes[1].grid()
axes[1].set_yscale('log')
axes[1].set_title('Min to Max')
axes[1].set_xlabel('Index')


for i in range(len(a)):
    if i % 2 == 0:
        axes[2].plot([i, i], [califica[i], califica[i]], color='blue', lw=2, marker='o')  # Plot the box from min to max
    else:
        axes[2].plot([i, i], [califica[i], califica[i]], color='red', lw=2, marker='o')
axes[2].set_xticks(np.arange(len(a)))
axes[2].set_xticklabels(range(len(a)),rotation=90)
axes[2].grid()
# axes[2].set_yscale('log')
axes[2].set_title('Cuanto mas alto mejor')
axes[2].set_xlabel('Index')



plt.show()
```
