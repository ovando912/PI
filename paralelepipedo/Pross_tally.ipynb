```python
import openmc
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd

plt.figure()

# Load the statepoint file 1
sp = openmc.StatePoint("sp1.h5")
tally = sp.get_tally(name="flux")
df = tally.get_pandas_dataframe()
df.columns = ["x", "y", "z", "nuclide", "score", "mean", "std.dev."]
plt.plot(df["z"], df["mean"])
plt.title("Flux vs. z")

# Load the statepoint file 2
sp = openmc.StatePoint("sp2.h5")
tally = sp.get_tally(name="flux")
df = tally.get_pandas_dataframe()
df.columns = ["x", "y", "z", "nuclide", "score", "mean", "std.dev."]
plt.plot(df["z"], df["mean"]*0.52)
plt.title("Flux vs. z")

plt.show()
```
