# Test 4: Prueba de tiempo en funcion de la cantidad de macrogrupos

## Read from file

```python
SurfaceSourceFile = kds.SurfaceSourceFile("surface_source.mcpl", domain={"w": [0, 1]})
df = SurfaceSourceFile.get_pandas_dataframe()
del SurfaceSourceFile
```

## Column order

```python
columns_order = ["x", "y", "ln(E0/E)"]
df = df[columns_order]
```

## Macro and micro bin sizes

```python
micro_bins = [150, 150, 150]
macro_bins = [100, 100, 100]
```

## Times

```python
tiempo_cumulative = []
tiempo_sample = []
tiempo_df = []
for macros in [25, 50,75,100]:
    macro_bins = [macros, macros, macros]
    print(macros)
    auxiliar_cumulative = []
    auxiliar_sample = []
    auxiliar_df = []
    for N in np.logspace(4,7,10):
        N = int(N)
        print(N)
        time_comulative, time_sample, time_df = tiempos_ejecucion(df, columns_order, macro_bins, micro_bins, N)
        auxiliar_cumulative.append(time_comulative)
        auxiliar_sample.append(time_sample)
        auxiliar_df.append(time_df)
    tiempo_cumulative.append(auxiliar_cumulative)
    tiempo_sample.append(auxiliar_sample)
    tiempo_df.append(auxiliar_df)
```

```python
Ns = np.logspace(4,7,10)

plt.figure()
plt.plot(Ns, tiempo_cumulative[0], label="25")
plt.plot(Ns, tiempo_cumulative[1], label="50")
plt.plot(Ns, tiempo_cumulative[2], label="75")
plt.plot(Ns, tiempo_cumulative[3], label="100")
plt.legend()
plt.xlabel("N")
plt.ylabel("Time")
plt.title("Cumulative time")
plt.show()

plt.figure()
plt.plot(Ns, tiempo_sample[0], label="25")
plt.plot(Ns, tiempo_sample[1], label="50")
plt.plot(Ns, tiempo_sample[2], label="75")
plt.plot(Ns, tiempo_sample[3], label="100")
plt.legend()
plt.xlabel("N")
plt.ylabel("Time")
plt.title("Sample time")
plt.show()

plt.figure()
plt.plot(Ns, tiempo_df[0], label="25")
plt.plot(Ns, tiempo_df[1], label="50")
plt.plot(Ns, tiempo_df[2], label="75")
plt.plot(Ns, tiempo_df[3], label="100")
plt.legend()
plt.xlabel("N")
plt.ylabel("Time")
plt.title("Dataframe time")
plt.show()
    
```
